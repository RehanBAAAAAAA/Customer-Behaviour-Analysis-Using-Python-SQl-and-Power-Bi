```python
# checking missing values
df.isnull().sum()
```

```python
#filling null values by calculating median by each category
df["Review Rating"] = df.groupby("Category")["Review Rating"].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# converting upper case to lower case and replacing space into "_"
# it is easy to read in sql it is to take refference without running code into error
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(" ","_")
df=df.rename(columns={"purchase_amount_(usd)":"purchase_amount"})
```

```python
df.columns
```

```python
# Creating a column age_group
labels=["Young Adult","Adult","Middle-aged","Senior"]
df["age_group"]=pd.qcut(df["age"],q=4,labels=labels)
```

```python
df[["age","age_group"]].head(10)
```

```python
# creating purchase_frequncey_days
# doing this to convert days into numbers so it will make easy to analyze
# eg weekly-14,annualy-365
frequency_mapping={"Fortnightly":14,
                   "Weekly":7,
                   "Monthly":30,
                   "Quarterly":90,
                   "Bi-Weekly":14,
                   "Annually":365,
                   "Every 3 Months":90}
df["purchase_frequency_days"]=df["frequency_of_purchases"].map(frequency_mapping)
```

```python
df[["purchase_frequency_days","frequency_of_purchases"]].head(10)
```
